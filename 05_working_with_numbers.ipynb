```python
# Basic numeric values
print(2)
print(-5)
print(2.45)
print(-3.44)

# Arithmetic operations
print(5 + 6.3)
print(5 - 6)
print(5 * 2)
print(10 / 2)
print((56 + 92) * 10)   # Using parentheses to control order of operations
print(10 % 3)           # Modulus operator (remainder)

# Storing numbers in variables
my_num = 69
print(my_num)

# Converting number to string for concatenation
print(str(my_num) + " my favourite number")

# Built-in numeric functions
num = -9
print(abs(num))         # Absolute value
print(pow(3, 2))        # Power function (3^2)
print(max(4, 6))        # Maximum value
print(min(4, 6))        # Minimum value
print(round(3.65))      # Rounding a number

# Importing math module functions
from math import *
print(floor(8.88))      # Rounds down
print(ceil(9.58))       # Rounds up
print(sqrt(36))         # Square root
```
