```python
# Basic built-in numeric functions
print(abs(-10))        # Absolute value
print(pow(2, 5))       # Power (2^5)
print(max(4, 9, 1))    # Maximum value
print(min(4, 9, 1))    # Minimum value
print(round(3.67))     # Rounding to nearest integer
print(sum([1, 2, 3, 4]))  # Sum of elements in a list
print(divmod(10, 3))   # Returns (quotient, remainder)

# Type conversion functions
print(int(3.9))        # Convert float to integer
print(float(5))        # Convert integer to float
print(complex(2, 3))   # Create complex number (2 + 3j)

# Number system conversions
print(bin(10))         # Convert to binary
print(oct(10))         # Convert to octal
print(hex(255))        # Convert to hexadecimal
```
