```python
# Creating a list
friends = ["Alice", "Bob", "Charlie", "David"]

print(friends)          # Print entire list
print(friends[0])       # Access first element
print(friends[-1])      # Access last element

# Modifying list element
friends[1] = "Mike"
print(friends)

# List slicing
print(friends[1:3])     # Print elements from index 1 to 2

# List functions
friends.append("John")  # Add element to end
friends.insert(1, "Tom")  # Insert element at specific position
friends.remove("Charlie")  # Remove element
friends.pop()           # Remove last element

print(len(friends))     # Length of list
print(friends)
```
