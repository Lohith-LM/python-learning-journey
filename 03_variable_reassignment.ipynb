```python
# Initial variable assignment
character_name = "Jhon"
character_age = "95"

# Building a sentence using string concatenation
print("There once was a man named " + character_name + " and")
print("he was " + character_age + " years old.")

# Reassigning variables with new values
character_name = "Mike"
character_age = "54"

# Output changes dynamically after reassignment
print("He really liked the name " + character_name + ",")
print("but didn't like being " + character_age + ".")
```
