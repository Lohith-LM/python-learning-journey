```python
# Using escape characters (\n for new line)
print("Girraffe\nAcademy")

# Using escape character to print double quotes inside a string
print("Hello everyone my name is \"LOHITH\" ")

xyz = "name-lohith"  # Storing a string in a variable

print(xyz)                   # Printing the stored string
print(xyz + " is cool")      # String concatenation
print(xyz.lower())           # Convert string to lowercase
print(xyz.upper())           # Convert string to uppercase
print(xyz.isupper())         # Check if entire string is uppercase
print(xyz.islower())         # Check if entire string is lowercase
print(xyz.upper().isupper()) # Chaining multiple string methods
print(len(xyz))              # Finding length of the string
print(xyz[0])                # Accessing character using index
print(xyz[6])
print(xyz.index("o"))        # Finding index position of a character
print(xyz.index("l"))
print(xyz.replace("name", "username"))  # Replacing part of a string
```
